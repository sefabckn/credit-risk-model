```python
import kagglehub

kagglehub.login()
```

```python
data_path = kagglehub.competition_download('home-credit-default-risk')
```

```python
import os
import shutil

raw_folder = os.path.join("data", "raw")
os.makedirs(raw_folder, exist_ok=True)

target_files = ["application_test.csv", "application_train.csv"]

for file in os.listdir(data_path):
    if file in target_files:
        shutil.copy(os.path.join(data_path, file), os.path.join(raw_folder, file))
```
